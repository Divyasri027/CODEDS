```python
data2=data.drop(['ANNUAL','Jan-Feb','Mar-May','Jun-Sep','Oct-Dec'],axis=1)#removes the columns
```

```python
data2['SUBDIVISION'].unique()#find the unique data
```

```python
data3=data2.loc[(data2.SUBDIVISION=="NAGA MANI MIZO TRIPURA")]#extracting the data of specific station
```

```python
data3
```

```python
data3.isna().sum()
```

```python
data4=data3.drop(['JAN','FEB','MAR','APR','MAY'],axis=1)#removes the specific columns
```

```python
data4
```

```python
data4['ANNUAL RAIN']=data3.apply(lambda row: row.JAN+row.FEB+row.MAR+row.APR+row.MAY+row.JUN+row.JUL+row.AUG+row.SEP+row.OCT+row.NOV+row.DEC,axis=1)#adding of columns
```

```python
data4
```

```python
data4['SWM']=data3.apply(lambda row: row.JUN+row.JUL+row.AUG+row.SEP,axis=1)
```

```python
data4
```

```python
data4['NEM']=data3.apply(lambda row: row.OCT+row.NOV+row.DEC,axis=1)
```

```python
data4
```

```python
data5=data4.drop(['JUN','JUL','AUG','SEP','OCT','NOV','DEC'],axis=1)
```

```python
data5
```

```python
cor=data5.corr()#correalation
cor
```

```python
import seaborn as sns #correlation graph
sns.heatmap(cor,vmax=1,vmin=-1,annot=True,linewidths=5,cmap='bwr')#blue-negative corr,#red-positive corr
```
